## Common Definitions

**Key Definitions:**
1. **Artificial Intelligence (AI):**
   - Development of computer algorithms to perform tasks usually associated with human intelligence.
   - Helps with tasks that are tedious, mundane, or require analyzing vast amounts of data.

2. **AI in Industry:**
   - Widely used across industries.
   - Encompasses a range of techniques, including:
     - **Machine Learning**
     - **Representation Learning**
     - **Deep Learning**
     - **Natural Language Processing (NLP)**

3. **Distinction Between AI Algorithms and Models:**
   - **AI Algorithms:**
     - Mathematical techniques designed for specific tasks (e.g., unsupervised or reinforcement learning).
     - Often developed by statisticians or mathematicians.
   - **AI Models:**
     - Computations that result from algorithms.
     - Take input values and produce output values.

4. **AI in Healthcare:**
   - **AI Solution:** An evaluated and validated AI model.
   - **AI Product:** A packaged AI solution for practical use.

## Why AI is needed in Healthcare?

- **Data Synthesis:**
  - Clinical data is overwhelming; AI helps synthesize vast amounts of medical information (clinical, genomic, social) to guide individualized patient care.
  - Example: AI can prioritize clinical actions (e.g., mammograms, vaccines) and highlight relevant clinical guidelines.

**AI in Biomedical Research:**
- **Data Mining:**
  - AI mines literature and clinical guidelines to assist in research.

  - AI tools perform tedious tasks without fatigue, reducing human errors.
  - Example: **Image labeling** in research.

**AI in Translational Research:**
- **Automatic Annotation:**
  - AI tackles annotation tasks like image labeling with high accuracy and speed, outperforming humans in certain diagnostic tasks.

**AI in Medical Practice:**
- **Patient Monitoring and Prioritization:**
  - AI solutions prioritize patients needing urgent care, such as in emergency departments, predicting deterioration risks.

- **Improving Clinician Engagement:**
  - AI aids data entry, reducing administrative burdens and freeing up time for doctor-patient interaction.

- **Enhancing Shared Decision-Making:**
  - AI analyzes structured and unstructured patient data, providing insights that assist physicians in understanding patients' values and goals.

**Conclusion:**

## Example of AI in Healthcare

**Expanding AI in Healthcare:**
- AI has long existed in healthcare but is now advancing through **medical imaging** and **other fields** like clinical practice, translational research, and biomedical research.

**AI in Medical Imaging:**
- **Example**: A deep learning algorithm from Stanford interpreted chest X-rays to detect 14 pathologies across 10 diseases.
  - It localized areas of interest on the image, performing as well or better than radiologists in certain pathologies.
  - **Efficiency**: Radiologists took 240 minutes to label 420 images; the AI algorithm took only 1.5 minutes.
  - Highlights how AI can reduce errors caused by human fatigue and improve diagnostic accuracy.

**AI in Translational Research:**
- **Example**: University of Toronto's AI models (support vector machines) classified proteins into drug/non-drug targets for cancer.
  - Key features: gene essentiality, mRNA expression, DNA copy number, mutations, and protein interaction networks.
  - **Discovery**: 53 novel cancer targets were identified across various sites.
  - Demonstrates AI's potential in **drug discovery**, moving beyond single-drug targets to uncover hundreds of potential cancer treatment targets.

**AI in Clinical Practice (Risk Stratification):**
- **Example**: Stanford's AI methods for stratifying prostate cancer patients based on risk of metastatic disease recurrence.
  - Combined **rule-based** and **convolutional neural networks** using structured and unstructured EHR data.
  - Tool adhered to clinical guidelines, synthesizing information to guide treatment decisions.
  
- **Example**: A tool developed at Stanford used for **hospital readmission risk stratification**.
  - Divided patients into low, intermediate, and high-risk groups, predicting the likelihood of readmission.
  - **Unsupervised clustering** revealed six high-risk patient subgroups.
  - Deployed into a statewide health information system for continuous patient monitoring during hospitalization.

**AI in Biomedical Research (Automated Experiments):**
- **Example**: Researchers at Brookhaven National Laboratory developed an **AI-driven autonomous experiment**.
  - AI explored multi-dimensional parameter spaces and improved model accuracy by focusing on areas of high uncertainty.
  - This **automated experimentation** reduced sampling error and improved knowledge over time, representing a breakthrough in basic biomedical research.

**AI in Autism Diagnosis:**
- **Example**: AI algorithms developed at Stanford used **home videos** to diagnose **autism spectrum disorder**.
  - Models assessed behaviors such as eye contact and social smile, achieving over 94% accuracy compared to expert raters.
  - Demonstrates AI's power in **diagnostics**, offering accurate and efficient solutions compared to traditional methods.

**Conclusion:**
- These examples illustrate how AI spans medical imaging, drug discovery, risk stratification, automated experimentation, and diagnostics. AI continues to enhance **care delivery, diagnostic accuracy, and efficiency**, leading to improved patient outcomes across the healthcare spectrum.

## Growth of AI in Healthcare

### Evaluating AI Solutions in Healthcare: A Systematic Framework

**Growing Interest in AI in Healthcare:**
- The field of AI in healthcare is experiencing exponential growth, with medical literature on AI applications expanding rapidly. In 2019, over 9,000 articles related to AI and healthcare were published, a fourfold increase over the previous decade.

**Challenges in Assessing AI Solutions:**
- Despite the hype, most AI models featured in research are not directly executable in clinical practice.
- Typical research focuses on:
  - Clinical problems justifying AI use.
  - Model architecture (e.g., CNN, machine learning models).
  - Accuracy or precision metrics (e.g., precision, recall, F-measure).
  - General thoughts on potential clinical application.
  
However, **key questions remain**:
- How effective are these solutions in real-world settings?
- Have they improved care or patient outcomes?
- How do we define a *good* AI solution?

**Evaluating AI Beyond Accuracy:**
- **Accuracy and precision metrics** alone are insufficient to evaluate an AI solution's true impact in healthcare.
- A good AI solution should not only be technically sound but also demonstrate real-world clinical effectiveness and improve patient care.

**Framework for Evaluating AI in Healthcare:**

1. **Clinical Utility:**
   - Does the AI model address a significant clinical problem?
   - How does it integrate into existing workflows?
   - Is it feasible for clinicians to use in real-time?
   
2. **Impact on Patient Outcomes:**
   - Has the AI model demonstrated improved patient outcomes (e.g., reduced mortality, better diagnostics)?
   - Are there case studies showing its impact on healthcare decisions or treatments?

3. **Scalability and Generalization:**
   - Can the AI model be generalized across different populations, hospitals, and health systems?
   - Does it perform well outside of the controlled research setting?

4. **Usability and Interpretability:**
   - Is the AI model interpretable by clinicians? Can they trust its predictions and recommendations?
   - Is the user interface intuitive for healthcare professionals to navigate?

5. **Ethical Considerations and Bias:**
   - Has the AI solution been tested for biases? Are there mechanisms in place to mitigate them?
   - Does the AI model ensure patient safety and comply with regulations?

6. **Cost-Effectiveness:**
   - Does the implementation of this AI model offer cost savings (e.g., reducing diagnostic errors or unnecessary tests)?
   - How does it compare to existing healthcare solutions in terms of cost and efficiency?

This framework expands beyond technical metrics like precision and recall, emphasizing the **practical impact of AI in healthcare settings**. A good AI solution should improve patient care, be scalable and interpretable, reduce costs, and address ethical issues effectively.

## Questions Answered by AI

### The Challenge of Clinical Utility in AI Models for Healthcare

**Focus on Available Labeled Datasets:**
- Many AI applications in healthcare, especially in predictive modeling, leverage readily available labeled datasets such as images or well-documented patient outcomes (e.g., readmissions or mortality).
- While this approach enables progress, it often leads researchers to focus on model performance (e.g., accuracy, C-statistics) without addressing the model's practical utility in clinical settings.

**Case Study: Hospital Readmission Prediction**
- In 2011, a systematic review analyzed prediction models for hospital readmissions:
  - About 8,000 citations were reviewed, and 30 studies met inclusion criteria.
  - Most models used **retrospective administrative data** and had poor discriminative power, with C-statistics under 0.8.
  - Only one study specifically focused on **preventable** readmissions, meaning most research efforts stopped at risk prediction without addressing clinical actionability.

**Clinical Utility Gap:**
- **Prediction alone is insufficient**: Identifying patients at high risk of readmission does not necessarily aid clinical decision-making if the model doesn’t provide actionable insights.
- **Clinical intuition**: Often, clinicians can already estimate which patients are at risk based on experience. The value of AI comes when it offers **new, actionable insights** beyond what is already known.
  
**Example of Actionable AI:**
- A model that not only predicts readmission but also identifies **modifiable factors** during the inpatient stay (e.g., prolonged prophylaxis post-surgery) could improve patient care by suggesting specific actions to reduce risk.
  
**Barriers to Effective AI Implementation:**
1. **Lack of Actionability**: Models often fail to suggest steps for mitigating predicted risks.
2. **No Integration with Clinical Workflow**: Most algorithms are developed without considering how they will fit into the busy clinical environment.
3. **Overlooking Clinical Feasibility**: Models might not account for the practical constraints clinicians face, such as time, resource availability, or the complexity of interventions.

### Conclusion:
AI models in healthcare should go beyond mere prediction. They need to provide **clinically useful** and **actionable insights** to improve patient outcomes. This requires a shift from focusing on technical model performance to considering the **clinical utility**, **feasibility**, and **impact** of AI solutions on real-world care delivery.

# How to Know if an AI model is Good?

## AI Output

### Evaluating an AI Solution in Healthcare: Moving Beyond Prediction

**Starting Point: Beyond Accuracy**
- A common pitfall in evaluating AI models in healthcare is stopping after achieving a high-performance metric like accuracy (e.g., a model predicting 30-day hospital readmissions with 94% accuracy). However, accuracy alone doesn’t provide insight into **clinical actionability**—what happens **next** after the prediction is made?

**Patient Timeline and Prediction Windows:**
- When evaluating an AI solution, it's important to think about a **patient's timeline** and the **window of observation** used for making predictions.
  - In the case of hospital readmissions, this observation window might be during the inpatient stay.
  - The prediction, such as the likelihood of readmission, is made at a set time point (e.g., at discharge), and you calculate a probability (e.g., 0.92).
  - After the prediction, there is a **window of action**—the time between the prediction and the event occurring—during which healthcare providers can take steps to mitigate the predicted event.

**Acute vs. Long-term Predictions:**
- The **window of action** can vary based on the type of prediction:
  - **Acute predictions** (e.g., patient deterioration within minutes or hours) require immediate action.
  - **Long-term predictions** (e.g., Alzheimer’s disease based on genetics) may not require action for years or even decades.
  - Regardless of the timeline, **early warning lead times** are valuable because they provide more opportunity for intervention.

**Example: Cardiac Arrest Prediction**
- Predicting cardiac arrest two minutes before the event gives clinicians little time to act, but predicting it two hours in advance provides more time to prevent the event.
  - This highlights the importance of **early predictions** that allow for timely actions.

### Broadening the Scope: From Prediction to Action
- Most current efforts in AI research focus on **data collection** and **model development** (e.g., data aggregation, model architecture, accuracy metrics). However, little attention is paid to the **action** side of the equation—what clinicians should do after receiving a prediction.

**Introducing the Outcome Action Pairing (OAP) Framework:**
- The **Outcome Action Pairing (OAP)** framework helps conceptualize the relationship between the **outcome** predicted by the AI model and the **action** that follows the prediction.
  
  1. **Outcome**: The purpose of the AI model, such as disease diagnosis, risk stratification, or event prediction.
     - Example: Predicting 30-day hospital readmissions.
  2. **Action**: The step that can be taken based on the outcome to improve patient care.
     - Example: A different treatment pathway based on risk, a scheduled follow-up to prevent readmission, or specific interventions triggered by a diagnosis.

**The Importance of Actionable AI Solutions:**
- For AI models to be useful in clinical practice, they must go beyond providing predictions to suggesting or enabling **clear, actionable steps** that improve patient outcomes.
- The **action** must align with clinical workflows and be feasible for healthcare teams to implement, ensuring the model’s integration into routine care.

### Conclusion: Moving Beyond Prediction
To evaluate AI solutions in healthcare:
1. Understand the **window of action** and whether the AI model provides enough lead time for meaningful intervention.
2. Use the **Outcome Action Pairing (OAP)** framework to ensure that the model’s predictions lead to actionable steps that improve clinical care.
3. Consider **who** will use the prediction and how it will fit into the healthcare setting, aiming for practical, actionable, and integrated AI solutions.

By focusing on the **action** component of AI evaluation, we can ensure that AI models are not only accurate but also impactful in improving patient outcomes and healthcare delivery.

## Think Beyond area under the curve

### Framework for Evaluating AI Solutions in Healthcare

To effectively evaluate AI solutions in healthcare, we can utilize a systematic framework that considers three key components: **utility, feasibility, and clinical impact**. Each component plays a crucial role in ensuring that AI solutions are not only innovative but also practical and beneficial for patients and healthcare providers.

#### 1. Utility
- **Definition**: Utility refers to the purpose and usefulness of the AI model in improving patient care and outcomes.
- **Key Questions**:
  - Does the AI solution provide predictions or insights that lead to a favorable change in patient care?
  - How does the model align with clinical needs and patient expectations?
  - Are the predictions actionable and relevant to the specific patient population?

#### 2. Feasibility
- **Definition**: Feasibility assesses the practical aspects of implementing the AI solution within the healthcare environment.
- **Key Questions**:
  - What resources (e.g., funding, technology, staff) are required to deploy the AI solution?
  - Is there sufficient IT infrastructure and support to integrate the AI model into existing workflows?
  - What training will staff need to effectively use the AI solution?
  - Are there any regulatory or compliance issues that need to be addressed?

#### 3. Clinical Impact
- **Definition**: Clinical impact evaluates the overall effect of the AI solution on healthcare practices, patient outcomes, and care standards.
- **Key Questions**:
  - What measurable improvements in patient outcomes can be attributed to the AI solution (e.g., reduced readmission rates, improved diagnosis accuracy)?
  - How does the AI model influence clinical decision-making and patient management?
  - Is there evidence from clinical trials or real-world studies demonstrating the effectiveness of the AI solution in practice?
  - How does the AI solution affect care standards and protocols?

### Implementation Considerations
- For an AI solution to be successfully integrated into healthcare, thoughtful consideration of these components is essential. Stakeholders, including clinicians, administrators, and patients, should be involved in the evaluation process to ensure that the solution addresses real-world challenges and meets the needs of all parties involved.

### Conclusion
This framework for evaluating AI solutions in healthcare emphasizes the importance of a holistic approach that goes beyond predictive accuracy. By systematically considering utility, feasibility, and clinical impact, we can better assess the potential of AI solutions to improve patient care, enhance operational efficiency, and ultimately transform healthcare delivery.

### Stakeholder Involvement in AI Solution Development in Healthcare

When developing AI applications in healthcare, engaging a diverse group of stakeholders is crucial to ensure the model addresses real-world needs and can be effectively integrated into clinical practice. Here’s a breakdown of key stakeholder groups and their roles:

#### 1. Knowledge Experts
- **Clinical Experts**: Physicians, nurses, and specialists who understand the clinical problems and can provide insights on patient care and treatment pathways.
- **Data Scientists**: Professionals skilled in data analysis, machine learning, and statistical methods who can help develop and validate the AI model.
- **Health Information and Technology Experts**: Individuals knowledgeable in healthcare IT systems who can facilitate the integration of AI solutions into existing workflows.
- **Implementation Experts**: Specialists who can advise on best practices for deploying and sustaining AI technologies in healthcare settings.

#### 2. Decision Makers
- **Hospital Administrators**: Leaders who can influence resource allocation, policy decisions, and overall strategy for adopting AI solutions.
- **Industrial Leadership**: Executives from technology companies or startups that develop AI tools and can provide insights on market trends and business viability.
- **Regulatory Agencies**: Bodies that ensure compliance with healthcare regulations and standards, which is vital for the adoption of AI technologies.
- **State and Federal Governments**: Policymakers who can provide funding, set regulations, and promote AI use in healthcare through initiatives and grants.

#### 3. End Users
- **Nurses and Physicians**: The primary users of AI tools who can provide feedback on usability, practicality, and the real-world applicability of the model.
- **Patients and Caregivers**: Individuals who ultimately benefit from AI solutions and whose perspectives are crucial for ensuring the model addresses their needs and preferences.
- **Laboratory Technicians**: Staff who work with diagnostic tools and technologies and can provide insights on integrating AI with laboratory processes.
- **Other End Users**: Any additional stakeholders who interact with or are affected by the AI solution, including administrative staff and health informatics professionals.

### Importance of a Multi-Disciplinary Team
A multi-disciplinary approach fosters collaboration among different stakeholders, ensuring that the AI solution is:
- Relevant to clinical needs.
- Feasible in terms of implementation and integration.
- Sustainable in real-world healthcare settings.
